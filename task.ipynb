# Imports

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
```

# Production (Sustainable)

```python
### Data Initialization
grayH_train = [95,85,75,65,50,35]
years = np.array([2020,2025,2030,2035,2040,2045]).reshape(-1,1)
grayH_test = [25]
greenH_train = [0,5,10,10,15,20]
greenH_test = [25]
blueH_train = [5,10,15,20,25,30]
blueH_test = [30]
turquoiseH_train = [0,0,0,5,10,15]
turquoiseH_test = [20]
test = np.array([2050]).reshape(1,1)
### Initializing Models
grayH = LinearRegression()
greenH = LinearRegression()
blueH = LinearRegression()
turquoiseH = LinearRegression()
### Training Models
grayH = grayH.fit(years,grayH_train)
y_pred = grayH.predict(test)
grayH_train.append(y_pred)
print(f"Gray Hydrogen RMSE is {((grayH_test-y_pred)**2)**0.5}")
greenH = greenH.fit(years,greenH_train)
y_pred = greenH.predict(test)
greenH_train.append(y_pred)
print(f"Green Hydrogen RMSE is {((greenH_test-y_pred)**2)**0.5}")
blueH = blueH.fit(years,blueH_train)
y_pred = blueH.predict(test)
blueH_train.append(y_pred)
print(f"Blue Hydrogen RMSE is {((blueH_test-y_pred)**2)**0.5}")
turquoiseH = turquoiseH.fit(years,turquoiseH_train)
y_pred = turquoiseH.predict(test)
turquoiseH_train.append(y_pred)
print(f"Turquoise Hydrogen RMSE is {((turquoiseH_test-y_pred)**2)**0.5}")

years=np.append(years,2050)
```

```python
print("\n\nGray Hydrogen Conventional\n\n")
grayH_train = np.array(grayH_train)
grayH_train_t = grayH_train*(0.1*180)
grayH_FullLoad = grayH_train_t*8760 
grayH_Electric = grayH_train_t*0.61
grayH_Water = grayH_train_t*2
grayH_TotalEnergy =grayH_train_t*34.4
grayH_NaturalGas = grayH_train_t*21
grayH_fuel = grayH_train_t*11.20

for i,n in enumerate(years):
    print(f"for year {n}\n\t Full Load Hours are {grayH_FullLoad[i]}\n\t Electric cost is {(grayH_Electric[i])}\n\t Water cost is {(grayH_Water[i])}\n\t Fuel cost is {grayH_fuel[i]}\n\t Natural Gas cost is {grayH_NaturalGas[i]}\n\t and Total energy cost is {grayH_TotalEnergy[i]}")
    print("\n================================\n")
    

print("\n\nGray Hydrogen New Plant\n\n")
grayH_train = np.array(grayH_train)
grayH_train_t = grayH_train*(0.1*180)
grayH_FullLoad = grayH_train_t*8760 
grayH_Electric = grayH_train_t*0.06
grayH_Water = grayH_train_t*2
grayH_TotalEnergy =grayH_train_t*28
grayH_NaturalGas = grayH_train_t*21
grayH_fuel = grayH_train_t*6.80

for i,n in enumerate(years):
    print(f"for year {n}\n\t Full Load Hours are {grayH_FullLoad[i]}\n\t Electric cost is {(grayH_Electric[i])}\n\t Water cost is {(grayH_Water[i])}\n\t Fuel cost is {grayH_fuel[i]}\n\t Natural Gas cost is {grayH_NaturalGas[i]}\n\t and Total energy cost is {grayH_TotalEnergy[i]}")
    print("\n================================\n")
    
    
print("\n\nGreen Hydrogen onsite\n\n")
greenH_train = np.array(greenH_train)
greenH_train_t = greenH_train*(0.1*180)
greenH_FullLoad = greenH_train_t*4000 
greenH_Electric = greenH_train_t*2.32
greenH_Water = greenH_train_t*1.59
greenH_TotalEnergy =greenH_train_t*8.4
greenH_NaturalGas = greenH_train_t*0
greenH_fuel = greenH_train_t*0

for i,n in enumerate(years):
    print(f"for year {n}\n\t Full Load Hours are {greenH_FullLoad[i]}\n\t Electric cost is {(greenH_Electric[i])}\n\t Water cost is {(greenH_Water[i])}\n\t Fuel cost is {greenH_fuel[i]}\n\t Natural Gas cost is {greenH_NaturalGas[i]}\n\t and Total energy cost is {greenH_TotalEnergy[i]}")
    print("\n================================\n")
    
print("\n\nGreen Hydrogen offsite\n\n")
greenH_train = np.array(greenH_train)
greenH_train_t = greenH_train*(0.1*180)
greenH_FullLoad = greenH_train_t*8760 
greenH_Electric = greenH_train_t*2.32
greenH_Water = greenH_train_t*1.59
greenH_TotalEnergy =greenH_train_t*8.4
greenH_NaturalGas = greenH_train_t*0
greenH_fuel = greenH_train_t*0

for i,n in enumerate(years):
    print(f"for year {n}\n\t Full Load Hours are {greenH_FullLoad[i]}\n\t Electric cost is {(greenH_Electric[i])}\n\t Water cost is {(greenH_Water[i])}\n\t Fuel cost is {greenH_fuel[i]}\n\t Natural Gas cost is {greenH_NaturalGas[i]}\n\t and Total energy cost is {greenH_TotalEnergy[i]}")
    print("\n================================\n")
    
print("\n\nBlue Hydrogen\n\n") 
blueH_train = np.array(blueH_train)
blueH_train_t = blueH_train*(0.1*180)
blueH_FullLoad = blueH_train_t*8760 
blueH_Electric = blueH_train_t*0.61
blueH_Water = blueH_train_t*2.00
blueH_TotalEnergy = blueH_train_t*34.4
blueH_NaturalGas = blueH_train_t*21
blueH_fuel = blueH_train_t*11.2

for i,n in enumerate(years):
    print(f"for year {n}\n\t Full Load Hours are {blueH_FullLoad[i]}\n\t Electric cost is {(blueH_Electric[i])}\n\t Water cost is {(blueH_Water[i])}\n\t Fuel cost is {blueH_fuel[i]}\n\t Natural Gas cost is {blueH_NaturalGas[i]}\n\t and Total energy cost is {blueH_TotalEnergy[i]}")
    print("\n================================\n")
print("\n\nTurquoise Hydrogen\n\n")
turquoiseH_train = np.array(turquoiseH_train)
turquoiseH_train_t = turquoiseH_train*(0.1*180)
turquoiseH_FullLoad = turquoiseH_train_t*4000 
turquoiseH_Electric = turquoiseH_train_t*2.32
turquoiseH_Water = turquoiseH_train_t*1.59
turquoiseH_TotalEnergy =turquoiseH_train_t*8.4
turquoiseH_NaturalGas = turquoiseH_train_t*0
turquoiseH_fuel = turquoiseH_train_t*0

for i,n in enumerate(years):
    print(f"for year {n}\n\t Full Load Hours are {turquoiseH_FullLoad[i]}\n\t Electric cost is {(turquoiseH_Electric[i])}\n\t Water cost is {(turquoiseH_Water[i])}\n\t Fuel cost is {turquoiseH_fuel[i]}\n\t Natural Gas cost is {turquoiseH_NaturalGas[i]}\n\t and Total energy cost is {turquoiseH_TotalEnergy[i]}")
    print("\n================================\n")
```

```python
plt.plot(years,grayH_train, label = 'Gray Hydrogen')
plt.plot(years,greenH_train, label = 'Green Hydrogen')
plt.plot(years,blueH_train, label = 'Blue Hydrogen')
plt.plot(years,turquoiseH_train, label = 'Turqouise Hydrogen')
plt.legend(loc='upper right')
plt.show()           
```
